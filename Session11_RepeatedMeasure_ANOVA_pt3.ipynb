```r
library(ggpubr)
library(readxl)
library(rstatix)
library(tidyverse)
```

```r
Learning_Dif = read_excel("D:\\Amar\\Codes_Data\\Learningdifficulty.xlsx")
```

```r
Learning_Dif
```

```r
Learning_Dif = Learning_Dif %>% gather (key = "time", value = "Learning_rate", pre_test, post_test, follow_up)
```

```r
Learning_Dif %>% group_by(Stimulation, Difficulty_Effect, time) %>% get_summary_stats(Learning_rate, type = "mean_sd")
```

```r
ggboxplot(Learning_Dif, x = "Stimulation", y = "Learning_rate", color = "time", facet.by = "Difficulty_Effect")
```

```r
Learning_Dif %>% group_by(Stimulation, Difficulty_Effect, time) %>% identify_outliers(Learning_rate)
```

```r
ggqqplot(Learning_Dif, "Learning_rate") + facet_grid(Stimulation +  Difficulty_Effect ~ time)
```

```r
Learning_Dif %>% group_by(Stimulation, Difficulty_Effect, time) %>% shapiro_test(Learning_rate)
```

```r
anova_test(data = Learning_Dif, dv = Learning_rate, wid = id, within = c(Stimulation, Difficulty_Effect, time))
```

```r
Learning_Dif %>% group_by(time, Stimulation) %>%  anova_test(dv = Learning_rate, wid = id, within = Difficulty_Effect) %>% adjust_pvalue (method = "bonferroni")
```

```r
Learning_Dif %>% group_by(time, Difficulty_Effect) %>%  anova_test(dv = Learning_rate, wid = id, within = Stimulation) %>% adjust_pvalue (method = "bonferroni")
```

```r
Learning_Dif %>% group_by(Stimulation, Difficulty_Effect) %>%  anova_test(dv = Learning_rate, wid = id, within = time) %>% get_anova_table() %>% adjust_pvalue (method = "bonferroni")
```
